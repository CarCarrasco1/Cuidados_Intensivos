### SHAPE

```python
dataframes= [admissions, callout, caregivers, chartevents, cptevents, d_cpt]

# Definir el atributo name para cada dataframe
admissions.name = "admissions"
callout.name = "callout"
caregivers.name = "caregivers"
chartevents.name = "chartevents"
cptevents.name = "cptevents"
d_cpt.name = "d_cpt"

# Inicializar una lista vacía para almacenar los resultados de shape
shapes = []

# Iterar sobre cada dataframe y aplicar la función shape
for i, df in enumerate(dataframes):
    shapes.append(df.shape)
    print(f"El dataframe {df.name} tiene forma {df.shape}")
```

### NULOS

NULOS POR DATAFRAME.

```python
dataframes= [admissions, callout, caregivers, chartevents, cptevents, d_cpt]

# Definir el atributo name para cada dataframe
admissions.name = "admissions"
callout.name = "callout"
caregivers.name = "caregivers"
chartevents.name = "chartevents"
cptevents.name = "cptevents"
d_cpt.name = "d_cpt"

# Iterar sobre cada dataframe e imprimir información sobre los valores nulos
for df in dataframes:
    total_null = df.isnull().sum().sum()
    print(f"El dataframe {df.name} tiene un total de {total_null} valores nulos.")
```
