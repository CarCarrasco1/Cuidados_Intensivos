```python
# FUNCION DE FILTRADO PARA EL DATAFRAME

def filtro(df, subject_id):
    # Filtrar los registros para el subject_id dado
    filtros = df[df['subject_id'] == subject_id]
    
    # Ordenar los registros por fecha de manera descendente
    final = filtros.sort_values('charttime', ascending=False)
    
    return final
```

```python
filtro(ml_df, 10011)
```

```python


# Creamos una columna con la fecha sin hora
ml_df['fecha'] = pd.to_datetime(ml_df['charttime']).dt.date

# Creamos una tabla pivote con una columna para cada label
base_ml = ml_df.pivot_table(index=['subject_id', 'fecha'], columns='label', values='valuenum', aggfunc='first')

# Reiniciamos el índice para que los campos 'subject_id' y 'fecha' vuelvan a ser columnas
base_ml = base_ml.reset_index()
```
