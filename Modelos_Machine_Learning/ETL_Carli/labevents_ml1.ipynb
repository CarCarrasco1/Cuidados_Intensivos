```python
import pandas as pd
import pymysql.cursors
import pymysql
```

```python
# Conexión a mi base de datos
connection = pymysql.connect(
    host='127.0.0.1',
    user='root',
    password='carla123',
    db='datawarehouse',
    charset='utf8mb4',
    cursorclass=pymysql.cursors.DictCursor
)

# Consulta SQL
sql = 'SELECT * FROM labevents'

# Creo un dataframe a partir de los datos de la tabla
labevents= pd.read_sql(sql, connection)

# Cierro la conexión
connection.close()

#Muestro mi dataframe
labevents.head()
```

```python
labevents.columns= labevents.columns.str.lower()
```

```python
diccionario = pd.read_csv("dataset/D_LABITEMS.csv")
diccionario.head()
```

```python
diccionario= diccionario.drop(['row_id', 'fluid', 'category', 'loinc_code'], axis=1)
```

```python
labevents = pd.merge(labevents, diccionario, on= 'itemid') 
```
