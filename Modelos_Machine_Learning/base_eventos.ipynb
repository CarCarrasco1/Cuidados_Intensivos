```python
patients_ml = pd.read_csv("dataset/PATIENTS.csv")
admissions_ml = pd.read_csv("dataset/ADMISSIONS.csv")
diagnoses_ml = pd.read_csv("dataset/DIAGNOSES_ICD.csv")
```

```python
#HAgo una lista con las columnas que necsito, y las separo armando nuevos dataframes.
columnas_d=['icd9_code','subject_id']
diagnoses_ml = diagnoses_ml.loc[:, columnas_d]
```

```python
columnas= ['subject_id', 'admittime', 
'deathtime',
'admission_type',
'ethnicity',
'diagnosis',
'hospital_expire_flag']
admissions_ml = admissions_ml.loc[:, columnas]
```

```python

columnas_p=['subject_id', 'gender', 'dob', 'dod'] 
patients_ml= patients_ml.loc[:, columnas_p]
```

```python
df_evento = admissions_ml.merge(patients_ml, on='subject_id').merge(diagnoses_ml, on='subject_id')
```
